Select Features and Target

```python
# Drop text columns
df = df.drop(['Entity', 'Code'], axis=1)

# Remove missing values
df = df.dropna()

print("Columns after cleaning:")
print(df.columns)
```

**Select Features and Target**

```python
# Define target column
target_column = 'Depressive disorders (share of population) - Sex: Both - Age: Age-standardized'

# Features (X)
X = df.drop(target_column, axis=1).values

# Target (y)
y = df[target_column].values

print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

**Normalize Data**

```python
import numpy as np
```

```python
X = (X - X.min(axis=0)) / (X.max(axis=0) - X.min(axis=0))

# reshape target
y = y.reshape(-1,1)

print("Normalization Done")
```

NumPy Neural Network (From Scratch)

STEP 1: Train/Test Split

```python
# Split data (80% train, 20% test)
split = int(0.8 * len(X))

X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

STEP 2: Initialize Parameters

```python
np.random.seed(42)

input_size = X_train.shape[1]
hidden_size = 32
output_size = 1

W1 = np.random.randn(input_size, hidden_size)
b1 = np.zeros((1, hidden_size))

W2 = np.random.randn(hidden_size, output_size)
b2 = np.zeros((1, output_size))
```

STEP 3: Activation Function

```python
def relu(z):
    return np.maximum(0, z)

def relu_derivative(z):
    return (z > 0).astype(float)
```

STEP 4: Forward Propagation

```python
def forward(X):
    global Z1, A1, Z2, A2

    Z1 = np.dot(X, W1) + b1
    A1 = relu(Z1)

    Z2 = np.dot(A1, W2) + b2
    A2 = Z2   # Linear output for regression

    return A2
```

STEP 5: Loss Function (MSE)

```python
def compute_loss(y_pred, y_true):
    return np.mean((y_pred - y_true) ** 2)
```

STEP 6: Backpropagation + Training

```python
learning_rate = 0.01
epochs = 500

for i in range(epochs):

    y_pred = forward(X_train)

    loss = compute_loss(y_pred, y_train)

    # Backpropagation
    dZ2 = y_pred - y_train
    dW2 = np.dot(A1.T, dZ2)
    db2 = np.sum(dZ2, axis=0, keepdims=True)

    dA1 = np.dot(dZ2, W2.T)
    dZ1 = dA1 * relu_derivative(Z1)
    dW1 = np.dot(X_train.T, dZ1)
    db1 = np.sum(dZ1, axis=0, keepdims=True)

    # Update
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2

    if i % 100 == 0:
        print("Epoch:", i, "Loss:", loss)
```

STEP 7: Test Accuracy

```python
test_pred = forward(X_test)
test_loss = compute_loss(test_pred, y_test)

print("Test Loss:", test_loss)
```

**PART 4 вЂ” PyTorch Model**

**STEP 1: Convert to Tensor**

```python
import torch
import torch.nn as nn
```

```python
X_train_t = torch.tensor(X_train, dtype=torch.float32)
y_train_t = torch.tensor(y_train, dtype=torch.float32)

X_test_t = torch.tensor(X_test, dtype=torch.float32)
y_test_t = torch.tensor(y_test, dtype=torch.float32)
```

**STEP 2: Define Model**

```python
class MentalNet(nn.Module):
    def __init__(self, hidden_size=8):
        super(MentalNet, self).__init__()
        self.hidden = nn.Linear(input_size, hidden_size)
        self.output = nn.Linear(hidden_size, 1)

    def forward(self, x):
        x = torch.relu(self.hidden(x))
        x = self.output(x)
        return x

model = MentalNet(hidden_size=8)
```

**STEP 3: Train**

```python
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

for epoch in range(500):

    y_pred = model(X_train_t)
    loss = criterion(y_pred, y_train_t)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch % 100 == 0:
        print("Epoch:", epoch, "Loss:", loss.item())
```
